```python
with t as (
	select
	 floor(count(case when (status = 'alive' and isinsider is false)
		  				then 1 else null end) * 0.90) as percent_90
from player
)
select
  	percent_90,
	case
		when percent_90 < (select amount from rations)
		then True else False end as rations
from t
```
