організатори хочуть визначити вразливих живих гравців, якими можна легко маніпулювати для наступної гри. Знайдіть усіх гравців, які живі, мають серйозні борги (борг > 400 000 000 вон), і вони або похилого віку (вік > 65), АБО мають порок азартних ігор без родинних зв’язків.

```python
select *
from player
where status = 'alive'
	and debt > 400000000
	and (age > 65 or (vice = 'Gambling' and has_close_family = false))
```

А тепер ваше наступне завдання:

організаторам потрібно розрахувати, скільки порцій їжі їм не давати, щоб створити потрібну кількість напруги. У таблиці обчисліть, якою кількістю пайків можна було б нагодувати 90% залишилися (живих) неінсайдерських гравців (з округленням у менший бік), а в іншому стовпці вкажіть, чи достатній поточний запас пайків. (Вірно чи хибно)

image.png

```python
with t as (
	select
	 floor(count(case when (status = 'alive' and isinsider is false)
		  				then 1 else null end) * 0.90) as percent_90
from player
)
select
  	percent_90,
	case
		when percent_90 < (select amount from rations)
		then True else False end as rations
from t
```
